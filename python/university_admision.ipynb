```python
#Pregunta: ¿Cuál es el promedio de las diferentes metricas por Rating de la Universidad)
#Primero vemos que valores toma la variable "University Rating"
Urating=adm_data["University Rating"].unique()
Urating
```

```python
Urating=np.sort(Urating)
print(Urating)
```

```python
#Veamos el promedio de las metricas para los casos con "University Rating=1"
w=adm_data["University Rating"]==1
tmp=adm_data[w]
tmp.mean()
```

```python
#Ahora lo hacemos para los diferente ratings
for rating in Urating:
  w=adm_data["University Rating"]==rating
  tmp=adm_data[w]
  print(f"Promedios para rating={rating}")
  print(tmp.mean())
```

```python
#Otra forma de responder a la pregunta usando la instrucción groupby
means_by_rating=adm_data.groupby(["University Rating"]).mean()
```

```python
means_by_rating
```
